```python
#Gender Demographics
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed
male_count=len(renamed_purchase_data.Players.loc[renamed_purchase_data["Gender"]=="Male"].unique())
print(f"Male Count:{male_count}")
female_count=len(renamed_purchase_data.Players.loc[renamed_purchase_data["Gender"]=="Female"].unique())
print(f"Female Count:{female_count}")
Other=len(renamed_purchase_data.Players.loc[renamed_purchase_data["Gender"]=="Other / Non-Disclosed"].unique())
print(f"Other/Non-Disclosed Count:{Other}")
male_percentage=(male_count/total_players*100) 
print(f"Male Percentage:{male_percentage}")
female_percentage=(female_count/total_players*100) 
print(f"Female Percentage:{female_percentage}")
Other_percentage=(Other/total_players*100) 
print(f"Other/Non-Disclosed Percentage:{Other_percentage}")
```

```python
purchase_count=purchase_data.SN.value_counts()
purchase_count_list=purchase_count.tolist()
players_list=purchase_count.index.tolist()
total_purchase_values=[purchase_data.Price[purchase_data["SN"]==player].sum() for player in players_list]
average_purchase_price=np.array(total_purchase_values)/np.array(purchase_count_list)
top_spenders={
            "Purchase Count":purchase_count_list,
            "Average Purchase Price":average_purchase_price,
            "Total Purchase Value":total_purchase_values
            }
top_spenders_df=pd.DataFrame(top_spenders, index=players_list)
top_spenders_df.index.name="SN"
top_spenders_df.sort_values("Total Purchase Value",ascending=False, inplace=True)
top_spenders_df.head()
```

```python
most_popular=purchase_data[["Item Name","Item ID","Price"]]
most_popular_group=most_popular.groupby(["Item ID","Item Name"])
purchase_count=most_popular_group["Item Name"].value_counts()
purchase_count1=purchase_count.values
total_purchase_value=most_popular_group.Price.sum()
average_price=total_purchase_value/purchase_count1
most_popular={"Purchase Count":purchase_count1,
             "Price":average_price,
             "Total Purchase Value":total_purchase_value
             }
most_popular_df=pd.DataFrame(most_popular)
most_popular_df.sort_values("Purchase Count", ascending=False,inplace=True)
most_popular_df.head()
```
